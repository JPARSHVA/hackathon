```python
import pandas as pd

df =pd.DataFrame({"a" :[],
                  "c" :[]})
df3 =pd.DataFrame({"a" :[],#Goal
                  "c" :[]})
df4 =pd.DataFrame({"a" :[],#card
                  "c" :[]})
a=pd.read_csv(filepath_or_buffer = "leister.csv")
a['c'].str.lower()
a= a.drop('b',axis=1)
```

```python
for x in range(90): 
  z= a.iloc[x,1]
  if z.find('card') != -1 or z.find('Goal') != -1 or z.find('free kick') != -1 or z.find('penalty') != -1 or z.find('off side') != -1:
    df2= a.iloc[x,:]
    df = df.append(df2)
df
```
